```python
import pygame,sys
from pygame.locals import *
from random import *
pygame.init()
FPS=10 #setting frames per second
fpsClock = pygame.time.Clock()
DISPLAYSURF = pygame.display.set_mode((1000,500)) #setting screen
sn = pygame.image.load('mario.png') #loading main body
sn = pygame.transform.scale(sn,(20,20))
food = pygame.image.load('food.png') #loading food
food = pygame.transform.scale(food,(20,20))
i=0
snake =[sn] #snake body list
x=[250] #list for x postion of snake body parts
y=[125] #list for x postion of snake body parts
WHITE = (255,255,255)
BLUE =  (  0,  0,255)
fnum = 'none' #initialized food decision
d=['right'] #direction

while True:
    DISPLAYSURF.fill(WHITE)
    # drawing boundary
    pygame.draw.line(DISPLAYSURF,BLUE,(10,10),(990,10))
    pygame.draw.line(DISPLAYSURF,BLUE,(990,10),(990,490))
    pygame.draw.line(DISPLAYSURF,BLUE,(990,490),(10,490))
    pygame.draw.line(DISPLAYSURF,BLUE,(10,490),(10,10))
    #wraping snape in screen
    if x[0]<10:
        x[0]=970
    elif x[0]>970:
        x[0]=10
    elif y[0]<10:
        y[0]=470
    elif y[0]>470:
        y[0]=10
    #printing food if there is no food
    if fnum=='none':
        foodx = randrange(10,970)
        foody = randrange(10,470)
        fnum = 'nnone'
    #snake eating food logic
    if fnum == 'nnone' and foodx-x[0]<20 and foodx-x[0]>-20 and foody-y[0]<20 and foody-y[0]>-20:
        snake.append(sn)
        i=i+1
        if d[i-1] == 'right':
            x.append(x[i-1]-20)
            y.append(y[i-1])
            d.append(d[i-1])
        elif d[i-1] == 'left':
            x.append(x[i-1]+20)
            y.append(y[i-1])
            d.append(d[i-1])
        elif d[i-1] == 'up':
            x.append(x[i-1])
            y.append(y[i-1]-20)
            d.append(d[i-1])
        elif d[i-1] == 'down':
            x.append(x[i-1])
            y.append(y[i-1]+20)
            d.append(d[i-1])
        fnum='none'
    #moving snake
    if d[0] == 'right':
        x[0] = x[0]+20
    elif d[0] == 'left':
        x[0] = x[0]-20
    elif d[0] == 'up':
        y[0] = y[0]-20
    elif d[0] == 'down':
        y[0] = y[0]+20
    #action handlers
    for event in pygame.event.get():
        # if we quit the game
        if event.type == QUIT:
            pygame.quit()
            sys.exit()
        # direction logic
        elif event.type == KEYDOWN and event.key == K_UP and d[0] != 'down'  :
            d[0] = 'up'
        elif event.type == KEYDOWN and event.key == K_DOWN and d[0] != 'up':
            d[0] = 'down'
        elif event.type == KEYDOWN and event.key == K_LEFT and d[0] != 'right':
            d[0] = 'left'
        elif event.type == KEYDOWN and event.key == K_RIGHT and d[0] != 'left':
            d[0] = 'right'
    # terminal condition(2 blocks have same coordinates)
    if fnum == 'nnone':
        for j in range(0,i):
            for k in range(j+1,i+1):
                if x[j]==x[k] and y[j]==y[k]:
                    pygame.quit()
                    sys.exit()
    #printing food and snake 
    DISPLAYSURF.blit(food,(foodx,foody))
    for j in range(0,i+1):
         DISPLAYSURF.blit(snake[j],(x[j],y[j]))
    #setting movement for next loop
    for j in range(i,0,-1):
        x[j]=x[j-1]
        y[j]=y[j-1]
        d[j]=d[j-1]
    pygame.display.update()
    fpsClock.tick(FPS)
```
